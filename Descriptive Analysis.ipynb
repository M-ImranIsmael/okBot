# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings(action='ignore')
```

# Overview of Datasets

```python
df = pd.read_csv("data_Syahirah_forshared.csv")
df.head()
```

# Data Distributions

```python
plt.subplot(1, 3, 1)
physical_dist = df["Class_Physicalhealth"].value_counts()
plt.title('Physical Health', size='16')
physical_dist.plot.bar(fontsize=14, rot=50, alpha=0.95, figsize=(15,10), color=sns.color_palette("deep"))
plt.ylabel('No of Students',size=14)
for i in range(2):
    plt.text(x=i, y=physical_dist[i], s=physical_dist[i], size=14, rotation=45, color='black')

plt.subplot(1, 3, 2)
plt.title('Relationship status', size='16')
relationships_dist = df["Relationship_status"].value_counts()
relationships_dist.plot.bar(fontsize=14, rot=50, alpha=0.95, figsize=(15,10), color=sns.color_palette("deep"))
for i in range(2):
    plt.text(x=i, y=relationships_dist[i], s=relationships_dist[i], size=14, rotation=45, color='black')
    
plt.subplot(1, 3, 3)
plt.title('Age Group', size='16')
age_dist = df["Age_group2"].value_counts()
age_dist.plot.bar(fontsize=14, rot=50, alpha=0.95, figsize=(15,10), color=sns.color_palette("deep"))
for i in range(2):
    plt.text(x=i, y=age_dist[i], s=age_dist[i], size=14, rotation=45, color='black')
```

# Relationships Analysis

```python
pd.crosstab(df.Class_LS, df.Financial_support).plot(kind="bar",
                                         figsize=(20,6),
                                         color=sns.color_palette("deep"),
                                         stacked=True,
                                         )
plt.title("Relationship between Life Satisfaction and Financial Support")
plt.ylabel("No of Students")
```

```python
pd.crosstab(df.Class_SE, df.Class_MW1).plot(kind="bar",
            figsize=(20,6),
            color=sns.color_palette("deep"),
            stacked=True
            )
plt.title("Relationship between Self Esteem Class and Mental WellBeing")

```

```python
pd.crosstab(df.Class_R, df.Class_MW1).plot(kind="bar",
                                         figsize=(20,6),
                                         color=sns.color_palette("deep"),
                                           stacked=True
                                         )
plt.title("Relationship between Resilience and Mental WellBeing")
```

```python
pd.crosstab(df.Class_SE, df.Class_R).plot(kind="bar",
                                         figsize=(20,6),
                                         color=sns.color_palette("deep"),
                                          stacked=True
                                         )
plt.title("Relationship between Self Esteem Class and Resilence")
```

# Correlation Matrix

```python
# Let's see how each variable correlate with each other
corr_matrix = df.corr()
fig, ax =plt.subplots(figsize=(15,10))
ax = sns.heatmap(corr_matrix,
                 annot =True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom+0.5,top-0.5)
ax.set_title('Correlation Matrix Between Each Variables')
```
